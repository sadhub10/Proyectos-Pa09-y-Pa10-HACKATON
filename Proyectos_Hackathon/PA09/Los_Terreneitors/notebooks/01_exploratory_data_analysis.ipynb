```python
# ===========================================
# NOTEBOOK 01 – Exploratory Data Analysis
# ===========================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar dataset
df = pd.read_csv("../../data/processed/cleaned_dataset.csv")

print("Primeras filas del dataset:")
df.head()
```

```python
# Información general del dataset
df.info()
```

```python
# Conteo de valores nulos
df.isnull().sum()
```

```python
# Distribución de etiquetas
plt.figure(figsize=(8,4))
sns.countplot(data=df, x="label", palette="viridis")
plt.title("Distribución de etiquetas")
plt.show()
```

```python
# Longitud del texto
df["text_length"] = df["text"].str.split().apply(len)

plt.figure(figsize=(8,4))
sns.histplot(df["text_length"], bins=50, kde=True)
plt.title("Distribución de longitud de textos")
plt.show()
```

```python
# Ejemplos por clase
for emotion in df["label"].unique():
    print("\n=== Ejemplos de clase:", emotion, "===\n")
    print(df[df["label"] == emotion]["text"].head(3).to_string(index=False))
```
